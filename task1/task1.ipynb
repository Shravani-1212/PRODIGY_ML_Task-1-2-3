```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# Read the data from CSV files
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
```

```python
print(train_data)
```

```python
print(test_data)
```

```python
print(train_data.columns)
```

```python
# Extracting features and target variable from training data
X_train = train_data[['GrLivArea', 'BedroomAbvGr', 'FullBath']].values
y_train = train_data['SalePrice'].values

# Extracting features from test data
X_test = test_data[['GrLivArea', 'BedroomAbvGr', 'FullBath']].values
```

```python
# Splitting training data into train and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
# Creating and training the linear regression model
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
# Making predictions on the validation set
predictions_val = model.predict(X_val)
```

```python
# Calculating Mean Squared Error (MSE) on validation set
mse_val = mean_squared_error(y_val, predictions_val)
print("Mean Squared Error on Validation Set:", mse_val)
```

```python
# Making predictions on the test set
predictions_test = model.predict(X_test)
```

```python
# Saving predictions to a CSV file
submission_df = pd.DataFrame({'Id': test_data['Id'], 'SalePrice': predictions_test})
submission_df.to_csv('submission.csv', index=False)
```
